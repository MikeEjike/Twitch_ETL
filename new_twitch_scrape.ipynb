```python
def get_streamer_data(html_element):
#     streamer_data = {}
    streamer_names = []
    streamer_live_view = []
    streamer_title = []
#     game_titles = [html_element]*25
    

        
    d = 0    
    for name in html_element.findAll('p', class_ = "tw-c-text-alt-2 tw-ellipsis"):
        d += 1        
        streamer_names.append(name.text)
        if d == 25:
            break
        
    e = 0
    for viewers in html_element.findAll('div', class_ = "tw-absolute tw-bottom-0 tw-left-0 tw-mg-1"):
        e += 1        
        streamer_live_view.append(viewers.text)
        if e == 25:
            break
        
    f = 0        
    for title_stream in html_element.findAll('h3', class_ = "tw-ellipsis tw-font-size-5"):
        f += 1        
        streamer_title.append(title_stream.text)
        if f == 25:
            break
    
    
#     streamer_data['Streamer'] = streamer_names
#     streamer_data['Stream Title'] = streamer_title
#     streamer_data['Live Views'] = streamer_live_view
# #     streamer_data['Game Playing'] = game_titles
#     streamer_df = pd.DataFrame.from_dict(streamer_data)
    
#     streamer_df['Trash'] = streamer_df['Live Views'].str.split(' ')
#     streamer_df['Live Views'] = streamer_df['Trash'].str.get(0)
#     streamer_df = streamer_df.drop(columns=['Trash'])
    
    return streamer_names, streamer_live_view, streamer_title
```

```python
def get_twitch_data(html_element):
    browser = Browser('chrome')
    url = 'https://www.twitch.tv/directory?sort=VIEWER_COUNT'
    browser.visit(url)
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    twitch_data = {}
    titles_clean = [] 
    viewer_count_clean = []
    
#     all_links= []
    
#     streamer_data = {}
    
    a = 0
    for title_div in html_element.findAll('span', class_ = "tw-c-text-alt"):
        a += 1
        if a == 25:
            break
        else:
            titles_clean.append(title_div.text)  
        
    b = 0        
    for view_div in html_element.findAll('p', class_ = "tw-c-text-alt-2"):
        b += 1
        if b == 25:
            break  
        else:
            viewer_count_clean.append(view_div.text)

#     for link in html_element.findAll('a', class_ = "tw-link tw-link--hover-underline-none tw-link--inherit")
#         all_links.append(link)

    c = 0    
    for title in titles_clean:
        c += 1
        if c == 25:
            break
        else:
            browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
#             browser.find_by_css('.tw-button-icon').first.click()
            browser.click_link_by_partial_text(title)
            time.sleep(2)
            try:
                browser.find_by_css('.tw-select-button').first.click()
                browser.click_link_by_partial_href('sort=VIEWER_COUNT')              
            except:
                print('NO "SORT BY"')
            html = browser.html
            soup = BeautifulSoup(html, 'html.parser')
            streamer_divs = soup.body
            get_streamer_data(streamer_divs)
            browser.visit(url)    
            time.sleep(2)
        print(c) 

            
    twitch_data['Games'] = titles_clean
    twitch_data['Total Live Views'] = viewer_count_clean
    twitch_df = pd.DataFrame.from_dict(twitch_data)
    
    twitch_df['Trash'] = twitch_df['Total Live Views'].str.split(' ')
    twitch_df['Total Live Views'] = twitch_df['Trash'].str.get(0)
    twitch_df = twitch_df.drop(columns=['Trash'])
    
    return twitch_df
```

```python
def top_streamers(url):
    streamer_list = {}
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    tables = soup.find_all("tbody")
    
    top_list = []
    bottom_list = []
    for top in tables[3].find_all("td"):
        top_list.append(top.text.strip())
    for bottom in tables[4].find_all("td"):
        bottom_list.append(bottom.text.strip())
    
    top_list_names = top_list[::2]
    bottom_list_names = bottom_list[::2]
    streamer_names_list = top_list_names + bottom_list_names
    
    top_list_views = top_list[1::2]
    bottom_list_views = bottom_list[1::2]
    streamer_views_list = top_list_views + bottom_list_views
    
    streamer_list['Streamers'] = streamer_names_list
    streamer_list['Live Views'] = streamer_views_list
    
    
    streamer_names = soup.body.findAll('td', class_ = 'cell-slot sm')
    
    for streamer_soup in streamer_names:
        stream = streamer_info(streamer_soup)
        views = streamer_views(streamer_soup)
        
        streamer_list.append(stream)
        streamer_list.append(views)
        
    return streamer_list
```

```python
def streamer_info(html_element):
    streamer_name_data = {}
#     table = html_element.find_all('table', class_='table')
#     table_df = pd.read_html(str(table))[0]
#     table_df.head()
    streamer_name_data['Streamer'] = html_element.find('td', class_ = 'cell-slot sm').text
#     streamer_data['Live Views'] = html_element.find('span').text
#     streamer_data['Live Views'] = html_element.find('td').find_next('td').text
    return streamer_name_data

def streamer_views(html_element):
    streamer_views_data = {}
#     table = html_element.find_all('table', class_='table')
#     table_df = pd.read_html(str(table))[0]
#     table_df.head()
#     streamer_data['Streamer'] = html_element.find('td', class_ = 'cell-slot sm').text
#     streamer_data['Live Views'] = html_element.find('span').text
    streamer_views_data['Live Views'] = html_element.find('td').find_next('td').text
    return streamer_views_data    
```
